```python
# Visualizations: Pairplot to visualize relationships between variables
plt.figure(figsize=(10, 8))
sns.pairplot(data, kind="scatter")
plt.suptitle('Pairwise Relationships between Variables', y=1.02)
plt.show()
```


### Visualization Findings

The scatter plots indicate relationships between the variables. Some takeaways:
- There seems to be a positive correlation between all subject scores, implying that if a student does well in one subject, they are likely to do well in others.
- Most students have scores between 60 and 80 in all subjects.
- There are a few outlier scores, including some zeros in "Pysics" and "Science", which might indicate either data entry errors or students who did not attend the test.
